```python
import cv2
frames = []

path = "/content/y2mate.com - Background  lửa cháy_480p.mp4"
cap = cv2.VideoCapture(path)
ret = True
while ret:
    ret, img = cap.read() 
    if ret:
        frames.append(img)
video = np.stack(frames, axis=0)
print(video)
```
